## Create custom images

```python
import matplotlib.pyplot as plt
```

```python
black = np.zeros((200,100),dtype = np.uint8)

plt.imshow(black, cmap = plt.cm.gray)
plt.title('black image')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
white = np.zeros((200,100)) + 255
white = np.uint8(white)

plt.imshow(white, cmap = plt.cm.gray, vmin = 0, vmax = 256)
plt.title("white image")
plt.xticks([]), plt.yticks([])
plt.show
```

```python
random_color_img = np.random.randint(0, high = 256, size = (150, 250, 3), dtype = np.uint8)

plt.imshow(random_color_img)
plt.title("random_color_img")
plt.xticks([]),plt.yticks([])
plt.show
```

Q1. Try to write a simple custom function to determine whether a given integer is odd or even number.

```python
def OddEven(x):
    "This function determines whether the integer is a odd or even number"
    if isinstance(x,int):
        if x % 2 == 0:
            if x == 0:
                statement = "is neither an odd or even number."
            else:
                statement = "is an even number."
        else:
            statement = "is an odd number."
    else:
        statement = "is not an integer."
    return print(f"{x} " + statement)    
```

```python
OddEven(0)
```

```python
OddEven(2)
```

```python
OddEven(3)
```

```python
OddEven("n")
```

Create an image as shown as the following with the help of Numpy and matplotlib modules. You can arbitrarily set the dimension of the image and white circular spot at the middle.

```python
# orginal black image
image = np.zeros((200,200),dtype = np.uint8)

# calculates the coordinates for the center
center_x = 200 // 2
center_y = 200 // 2

# assigning a radius for the circle
radius = 25
# to iterate over every pixel in the image(200,200)
for x in range(200):
    for y in range(200):
        if (x - center_x) ** 2 + (y - center_y) ** 2 <= radius ** 2: # the calculate the pixels forming the circle inside the center of the image
            image[x,y] = 255 # change the values of the pixels inside the circle to 255, which is white
# the image now has a white circle in the center             
plt.imshow(image, cmap = plt.cm.gray)
plt.title("Black image with white circle in the center")
plt.xticks([]), plt.yticks([])
plt.show
```
